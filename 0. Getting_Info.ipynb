```python
def toString(s):
    if s is None:
        return "None"
    if isinstance(s, (str)):
        return s
    return repr(int(s))


def chech(s):
    if isinstance(s, (str)):
        return s


data = []

with open_workbook('FQV.xlsb') as wb:
    for sheetname in wb.sheets:
        print(sheetname)
        f = open(sheetname + ".csv", "w+")
        with wb.get_sheet(sheetname) as sheet:
            for row in sheet.rows():
                values = [r.v for r in row]
                f.write('~'.join(map(toString, values)) + "\n")
```
